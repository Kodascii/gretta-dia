## Exercice: [Spirale](https://python.sdv.univ-paris-diderot.fr/07_fichiers/#773-spirale-exercice)

```python
import os
import math
import matplotlib.pyplot as plt

data_path = "./res/spirale.dat"

def calc_coordinates():
    angle = 0
    r = 0.1
    step = 0.1

    with open(data_path, 'wt') as data:
        while angle < 16 * math.pi:
            x = round(r * math.cos(angle), 4)
            y = round(r * math.sin(angle), 4)

            angle += step
            r += step
            data.write(f'{x} {y}\n')

def plot():
    x = []
    y = []
    with open(data_path, 'rt') as f_in:
        for line in f_in:
            coords = line.split()
            x.append(float(coords[0]))
            y.append(float(coords[1]))

    plt.figure(figsize=(8,8))
    mini = min(x+y) * 1.2
    maxi = max(x+y) * 1.2
    plt.xlim(mini, maxi)
    plt.ylim(mini, maxi)
    plt.plot(x, y)
    plt.savefig('./res/spirale.png')

calc_coordinates()
plot()
```

## Pandas DataFrame to CSV

```python
import numpy as np
import pandas as pd

path = './res/panda.csv'
two_pi = 2 * np.pi

data = {
    'A': np.random.randint(30),
    'B': [np.cos(y) for y in np.arange(0, two_pi, two_pi/30)],
    'C': np.random.rand(30)
}

df = pd.DataFrame(data)
df.to_csv(path)
```
