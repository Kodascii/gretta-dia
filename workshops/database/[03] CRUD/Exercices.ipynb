## [Exercice 01](https://openclassrooms.com/fr/courses/6971126-implementez-vos-bases-de-donnees-relationnelles-avec-sql/7142248-inserez-des-donnees-dans-votre-bdd#/id/r-8011785)
> **Consigne** : *Créer un nouvel aliment. En voici les données nutritionnelles :*

| nom            | marque   | calories | sucre | graisses | proteines | bio   |
|----------------|----------|----------|-------|----------|-----------|-------|
| haricots verts | Monoprix | 25       | 3     | 0        | 1,7       | FASLE |

````python
%%sql
INSERT INTO `aliment` (`nom`, `marque`, `calories`, `sucre`, `graisses`, `proteines`, `bio`)
VALUES ('haricots verts', 'Monoprix', 25, 3, 0, 1.7, false);

SELECT * FROM aliment WHERE `id` >= 15;
````

## [Exercice 03](https://openclassrooms.com/fr/courses/6971126-implementez-vos-bases-de-donnees-relationnelles-avec-sql/7142260-mettez-a-jour-les-donnees-de-votre-bdd#/id/r-7978405)
> **Consigne** : *Nous avons oublié de préciser le type de pomme vendue à Intermarché. Il s'agit d'une pomme golden. Comment feriez-vous pour rectifier le tir ?*

````python
%%sql
UPDATE `aliment` SET `nom` = 'pomme golden', `marque` = 'Intermarché' WHERE `id` = 1;

SELECT * FROM `aliment` WHERE `id` <= 5;
````

## [Exercice 04](https://openclassrooms.com/fr/courses/6971126-implementez-vos-bases-de-donnees-relationnelles-avec-sql/7142269-extrayez-des-informations-specifiques-de-votre-bdd#/id/r-7980425)
> **Consignes** : *Imaginons que je vous demande tous les aliments qui ne sont pas bio dans la base, classés par ordre décroissant de contenance en protéines.*

````python
%%sql
SELECT * FROM `aliment`
WHERE `bio` = false
ORDER BY `proteines` DESC;
````
